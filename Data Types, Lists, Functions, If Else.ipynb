# Data Types

```python
#integer - this is a data type to describe numbers
my_integer=1
print(my_integer)
print(type(my_integer))

#float - this is a data type to describe numbers
my_float = 3.14
print(my_float)
print(type(my_float))

#string
my_string = "Shrek"
print(my_string)
print(type(my_string))

#Boolean - True/False
my_Boolean = 1<2
another_Boolean = 2==3
print(my_Boolean)
print(another_Boolean)
print(type(my_Boolean))


#Lists - Can be used to describe vectors, or anything really
my_list = [1,2,"apple"]
print(my_list)
print(type(my_list))
print (my_list*2)
print(my_list[2])
```

# Functions

```python
y=3.2 #give variable a value

#instead, to define a function
def PrintAbs(x):
    print("The absolute value of",x,"is",(x**2)**0.5)

def abs(x):
    return (x**2)**0.5 #use return if you want the function to retain a certain value after being executed. That way it can be used as number or variable

#Can make recursvie functions (not ususally a best practice, better to use loops...as we will learn about next time)
def recursive(x):
    if x<3:
        print(x)
        recursive(x+1)
    else: print("recursion complete")
```

```python
PrintAbs(-6.28)

print(2*abs(-3.14))

recursive(-8)
```
